```python
# Crawl Data

filename = 'MBG.csv'
search_keyword = 'Makan Bergizi Gratis since:2025-06-25 until:2025-06-30  lang:id'
limit = 1000

!npx -y tweet-harvest@2.6.1
-o "{filename}" -s "{search_keyword}"
--tab "LATEST" -l {limit} --token {twitter_auth_token}
```

```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# Display the DataFrame
display(df)
```
